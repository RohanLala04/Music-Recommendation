```python
X_train, X_test = train_test_split(df, test_size = 0.75, random_state=10)
```

```python
X_train.head(30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(X_train['Artists'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim[:4, :4]
```

```python
#Constructing a reverse map of indices and song titles
indices = pd.Series(X_train.index, index=X_train['name']).drop_duplicates()
artists = pd.Series(X_train.index, index=X_train['Artists']).drop_duplicates()
pd.DataFrame(artists).head()
```

```python
# Function that get song recommendations based on the cosine similarity score of artist names
def genre_recommendations(name):
    idx = indices[name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    song_indices = [i[0] for i in sim_scores]
    return artists.iloc[song_indices]
```
